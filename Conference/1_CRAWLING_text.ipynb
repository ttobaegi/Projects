```python
import time
import re
import pandas as pd

class Crawler:
    def __init__(self, urls, path):
        self.driver = webdriver.Chrome(path)
        self.urls = urls
        self.dataframe=None
        self.erridx=[]
        self.title_list=[]
        self.author_list=[]
        self.date_list=[]
        self.text_list=[]
        self.buyer_list=[]
        self.isbm_list=[]
    
    def crawl(self):
        col=['제목','지은이','출판년도','줄거리','여1','남1','여2','남2','여3','남3','여4','남4','여5','남5','여6','남6','ISBM']
        for i in range (12):
            self.buyer_list.append([])
        for url in self.urls:
            tmp_vars=[]
            self.driver.get(url)
            self.driver.implicitly_wait(3)
            self.title_list.append(self.driver.find_element_by_class_name('Ere_bo_title').text)
            tl=self.driver.find_elements_by_class_name('Ere_sub2_title')
            self.author_list.append(tl[1].text)
            day=tl[0].text
            for i in range(1,len(tl)):
                day=day.replace(tl[i].text,"")
            day=day.replace(' (지은이)','').replace(' (옮긴이)','').replace(',','')
            self.date_list.append(day)
            tmp_isbm = self.driver.find_element_by_class_name('conts_info_list1').find_elements_by_tag_name('li')
            self.isbm_list.append(tmp_isbm[len(tmp_isbm)-1].text.replace('ISBN : ',''))
            self.scroll()
            text=''
            li = self.driver.find_elements_by_xpath("//div[@id='Underline3Updates']//div[@class='Rconts2']")
            if len(li)==0:
                ii=self.driver.find_elements_by_xpath("//div[@class='Ere_prod_mconts_box']")
                for i in ii:
                    try:
                        if i.find_element_by_class_name('Ere_prod_mconts_LS').text == '책소개':
                            text=self.dore(i.find_element_by_class_name('Ere_prod_mconts_R').text)
                            break
                    except:
                        text=''
            else:
                cnt=1
                for i in li:
                    try:
                        i.find_element_by_xpath("//span[@id='u3_"+str(cnt)+"']").click()
                        time.sleep(0.01)
                        text += self.dore(i.find_element_by_xpath("//span[@id='u3_"+str(cnt)+"_more']").text[:-3])
                    except:
                        text += self.dore(i.find_element_by_xpath("//span[@id='u3_"+str(cnt)+"']").text[:-3])   
                    cnt+=1
            
            self.text_list.append(text)
            
            buyerli = self.driver.find_elements_by_xpath("//div[@class='buyer_area_a']//div[@class='per']")
            
            for i in range(len(buyerli)):
                self.buyer_list[i].append(buyerli[i].text)
            #print(title_list,author_list,date_list,text_list,buyer_list)
        self.dataframe=pd.DataFrame({col[0]:self.title_list,col[1]:self.author_list,col[2]:self.date_list,col[3]:self.text_list,col[4]:self.buyer_list[0],col[5]:self.buyer_list[1],col[6]:self.buyer_list[2],col[7]:self.buyer_list[3],col[8]:self.buyer_list[4],col[9]:self.buyer_list[5],col[10]:self.buyer_list[6],col[11]:self.buyer_list[7],col[12]:self.buyer_list[8],col[13]:self.buyer_list[9],col[14]:self.buyer_list[10],col[15]:self.buyer_list[11],col[16]:self.isbm_list})
    
    def scroll(self):
        for i in range(1,11):
            self.driver.execute_script("window.scrollTo(document.body.scrollHeight/11*"+str(i-1)+", document.body.scrollHeight/11*"+str(i)+");")
            time.sleep(0.5)
    
    
    def dore(self,text):
        return re.sub('[^a-zA-Z|ㄱ-ㅣ가-힣|0-9]+',' ', text)
        
    def get_df(self):
        return self.dataframe
    
```

```python
class Crawler:
    def __init__(self, urls, path):
        self.driver = webdriver.Chrome(path)
        self.urls = urls
        self.dataframe=None
        self.erridx=[]
        self.title_list=[]
        self.author_list=[]
        self.date_list=[]
        self.text_list=[]
        self.buyer_list=[]
        self.isbm_list=[]
    
    def crawl(self):
        col=['제목','지은이','출판년도','줄거리','여1','남1','여2','남2','여3','남3','여4','남4','여5','남5','여6','남6','ISBM']
        for i in range (12):
            self.buyer_list.append([])
        for url in self.urls:
            tmp_vars=[]
            self.driver.get(url)
            self.driver.implicitly_wait(3)
            self.title_list.append(self.driver.find_element_by_class_name('Ere_bo_title').text)
            tl=self.driver.find_elements_by_class_name('Ere_sub2_title')
            self.author_list.append(tl[1].text)
            day=tl[0].text
            for i in range(1,len(tl)):
                day=day.replace(tl[i].text,"")
            day=day.replace(' (지은이)','').replace(' (옮긴이)','').replace(',','')
            self.date_list.append(day)
            tmp_isbm = self.driver.find_element_by_class_name('conts_info_list1').find_elements_by_tag_name('li')
            self.isbm_list.append(tmp_isbm[len(tmp_isbm)-1].text.replace('ISBN : ',''))
            self.scroll()
            text=''
            li = self.driver.find_elements_by_xpath("//div[@id='Underline3Updates']//div[@class='Rconts2']")
            if len(li)==0:
                ii=self.driver.find_elements_by_xpath("//div[@class='Ere_prod_mconts_box']")
                for i in ii:
                    try:
                        if i.find_element_by_class_name('Ere_prod_mconts_LS').text == '책소개':
                            text=self.dore(i.find_element_by_class_name('Ere_prod_mconts_R').text)
                            break
                    except:
                        text=''
            else:
                cnt=1
                for i in li:
                    try:
                        i.find_element_by_xpath("//span[@id='u3_"+str(cnt)+"']").click()
                        time.sleep(0.01)
                        text += self.dore(i.find_element_by_xpath("//span[@id='u3_"+str(cnt)+"_more']").text[:-3])
                    except:
                        text += self.dore(i.find_element_by_xpath("//span[@id='u3_"+str(cnt)+"']").text[:-3])   
                    cnt+=1
            
            self.text_list.append(text)
            
            buyerli = self.driver.find_elements_by_xpath("//div[@class='buyer_area_a']//div[@class='per']")
            
            for i in range(len(buyerli)):
                self.buyer_list[i].append(buyerli[i].text)
            #print(title_list,author_list,date_list,text_list,buyer_list)
        self.dataframe=pd.DataFrame({col[0]:self.title_list,col[1]:self.author_list,col[2]:self.date_list,col[3]:self.text_list,col[4]:self.buyer_list[0],col[5]:self.buyer_list[1],col[6]:self.buyer_list[2],col[7]:self.buyer_list[3],col[8]:self.buyer_list[4],col[9]:self.buyer_list[5],col[10]:self.buyer_list[6],col[11]:self.buyer_list[7],col[12]:self.buyer_list[8],col[13]:self.buyer_list[9],col[14]:self.buyer_list[10],col[15]:self.buyer_list[11],col[16]:self.isbm_list})
    
    def scroll(self):
        for i in range(1,11):
            self.driver.execute_script("window.scrollTo(document.body.scrollHeight/11*"+str(i-1)+", document.body.scrollHeight/11*"+str(i)+");")
            time.sleep(0.5)
    
    
    def dore(self,text):
        return re.sub('[^a-zA-Z|ㄱ-ㅣ가-힣|0-9]+',' ', text)
        
    def get_df(self):
        return self.dataframe
    
```
