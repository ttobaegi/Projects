```python
for i in range (1,7):
    response = requests.get('https://www.aladin.co.kr/shop/common/wbest.aspx?BestType=YearlyBest&BranchType=1&CID=55889&Year=2018&page='+str(i)+'&cnt=300&SortOrder=1')
    soup = BeautifulSoup(response.text, 'html.parser')
    for box in soup.select('div[class=ss_book_box]'):
        img_url[box.select('a[class=bo3]')[0].text] = box.select('img[class=i_cover]')[0].get("src")
```
