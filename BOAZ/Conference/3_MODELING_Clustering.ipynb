##MODELING

```python
from gensim.models.doc2vec import TaggedDocument
class Doc2VecCorpus:
    def __init__(self, tts):
        self.tts = tts
    def __iter__(self):
        for name,sentence in tts:
            if(len(sentence)>0):
                yield TaggedDocument(
                    words = sentence,
                    tags = [name])
```

```python
tts = []
for tag,ss in zip(df['name'], sentence):
    tts.append([tag, komo.nouns(ss)])
tts=tts*20
```

```python
doc2vec_corpus2 = Doc2VecCorpus(tts)
from gensim.models import Doc2Vec

doc2vec_model2 = Doc2Vec(doc2vec_corpus2)
```

```python
for idx, doctag in sorted(doc2vec_model2.docvecs.doctags.items(), key=lambda x:x[1].offset):
    print(idx, doctag)
```
